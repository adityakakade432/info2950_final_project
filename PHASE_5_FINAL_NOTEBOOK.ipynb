## Hypotheses & Analyses

### Hypothesis 1: A Higher Borrower's Obligation Implies That a Country is More Likely to be a Developing Economy

We will now plot a scatterplot to distinguish the developed and developing economies' borrowers obligation to see whether a higher borrower's obligation would most likely be among developed or developing economies. To do so, we will use binary values 1 and 0, 1 being developed and 0 being a developing economy.

```python
#plotting the data prior to the regression analysis, to get a visual
#understanding of borrower's obligation  in relation to developed/developing

plt.figure(figsize = (10, 6))
sns.scatterplot(y = gdp_debt_csv['developed'], x = gdp_debt_csv['Borrower\'s Obligation'])
plt.ylabel('Developed Binary Value')
plt.xlabel('Borrower\'s Obligation (Billions of $)')
plt.legend()
```

This scatter plot shows the relationship between borrower's obligation (in billions of $) on the x-axis and a binary variable for whether a country is developed (1 for developed, 0 for developing) on the y-axis. Most of the points are clustered at the lower end of the borrower's obligation spectrum, indicating that both developed and developing countries have obligations in this range. However, the developed countries are spread across the entire range of obligations, while developing countries are not present in the higher obligation values.


This visual suggests that while higher obligations are associated with developed countries, there is not an exclusive relationship, as developed countries are also present in the lower obligation values. This provides an initial visual indication that higher borrower's obligations might be associated with developed countries, but the relationship is not definitive. Thus, we will conduct a logistic regression to see whether the relationship between borrower's obligation and the status of it being developed or developing is significant. 


```python
#fitting model and looking at coefficient
y = gdp_debt_csv['developed']
X = gdp_debt_csv[['Borrower\'s Obligation']]
X_sm = sm.add_constant(X)
b_obl_model = sm.Logit(y,X_sm).fit()
print(b_obl_model.summary())
```

The output is from a logistic regression model, with 'developed' as the dependent binary variable and 'Borrower's Obligation' as the independent variable. The model's coefficient for 'Borrower's Obligation' is negative, and the p-value is 0.088, which is slightly above the common alpha level of 0.05 for statistical significance.
The coefficient of about 0.28 suggests that as borrower's obligation increases by one unit, the likelihood of a country being developed decreases by about 72 percent. However, the p-value indicates that this relationship is not statistically significant at the 5% level, meaning we cannot confidently reject the null hypothesis that there is no relationship between borrower's obligation and the likelihood of being a developed country.  

Now that we know from the logistic regression output that we should reject our null hypothesis 1 (there is no relationship). We want to see specific behavior of GDP, as it is an indicator for whether a country is developed or developing. By including the pattern across years, we can have a more holistic view of its interaction with the borrower's obligation


This heatmap shows the correlation between GDP by year (each row represents a different year's GDP) and borrower's obligation by year (each column represents a different year's obligation). Red shades indicate positive correlation, and blue shades indicate negative correlation. The heatmap provides a more nuanced view, showing varying degrees of correlation between GDP and borrower's obligation across different years. As we assume that GDP is a proxy for development, this heatmap suggests that the relationship between borrower's obligation and a country's development status is complex and varies over time, which is important for interpreting the hypothesis.

In summary, for hypothesis 1, while there is some visual indication that higher borrower's obligations might be associated with developed countries, the logistic regression analysis does not find a statistically significant relationship.
```python
ip_debt_csv = pd.read_csv("./Int_Prop_Debt_Combined.csv")
print(ip_debt_csv.head(2))
```

In this step we are calculating the Intellectual Property Charges over the specified years and creating a new column (‘avg’) to simplify analysis. Instead of working with individual numbers for different countries, creating the column allows us to have a single representative value for all of the developed countries. It also makes it more readable with concise numbers of the Intellectual Property Charges. After the previous step, we are moving on to GDP. We are calculating average GDP from all the developed countries and creating a new column (‘avg’) for simplified analysis. As explained in the previous section, having the average instead of individual numbers for each country makes the data easier to handle and interpret. The ‘avg’ column also helps with the data being easily integrated into different types of analysis such as regression models. This is because it is a comprehensive variable. 


```python
data_columns = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008',
       '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017',
       '2018', '2019', '2020', '2021', '2022']

data_columns_gdp = ['2000GDP', '2001GDP', '2002GDP',
       '2003GDP', '2004GDP', '2005GDP', '2006GDP', '2007GDP', '2008GDP', '2009GDP', '2010GDP', '2011GDP',
       '2012GDP', '2013GDP', '2014GDP', '2015GDP', '2016GDP', '2017GDP', '2018GDP', '2019GDP', '2020GDP',
       '2021GDP', '2022GDP']

ip_avg = ip_debt_csv[data_columns].mean(axis = 1)
ip_debt_csv['avg'] = ip_avg
print(ip_debt_csv.head())

gdp_avg = gdp_debt_csv[data_columns_gdp].mean(axis = 1)
gdp_debt_csv['avg'] = gdp_avg
print(gdp_debt_csv.head())
```

In summary, we imported the data, calculated the average intellectual property charges across specified years for developed countries, repeated this process for GDP, and then fit a linear regression model to determine the relationship between the variables. Now, we will conduct a linear regression for both developing and developed countries to see if Number of Years Since 2000 has a significant impact on either category's Intellectual Property Charges.

```python
#OLS for Developed Countries
y = ip_avg_deved
X = years_since_2000
X_sm = sm.add_constant(X)
ip_years_model_deved = sm.OLS(y,X_sm).fit()
print("Developed Country Summary:")
print(ip_years_model_deved.summary())
```

```python
#OLS for Developing Countries
y = ip_avg_deving
X = years_since_2000
X_sm = sm.add_constant(X)
ip_years_model_deving = sm.OLS(y,X_sm).fit()
print("Developing Country Summary:")
print(ip_years_model_deving.summary())
```
